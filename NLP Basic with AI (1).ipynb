```python
#Chunking
import nltk
from nltk.tokenize import word_tokenize
lotr_quote = "It's a dangerous business, Frodo, going out your door."
words_in_lotr_quote = word_tokenize(lotr_quote)
words_in_lotr_quote

```

```python
nltk.download("averaged_perceptron_tagger")
lotr_pos_tags = nltk.pos_tag(words_in_lotr_quote)
lotr_pos_tags
```

```python
grammar = "NP: {<DT>?<JJ>*<NN>}"
#Start with an optional (?) determiner ('DT')
#Can have any number (*) of adjectives (JJ)
#End with a noun (<NN>) 
chunk_parser = nltk.RegexpParser(grammar)
tree = chunk_parser.parse(lotr_pos_tags)
```

```python
#from IPython.display import display
display(tree)
```

```python
#Name Entity Recognition
quote='Barack Obama visited Microsoft in Seattle.'
tokens = word_tokenize(quote)
pos_tags = nltk.pos_tag(tokens)

# 4. Perform NER (creates an NLTK Tree structure)
ner_tree = nltk.ne_chunk(pos_tags)
for subtree in ner_tree:
    if isinstance(subtree, nltk.Tree):
        # Extract the label (e.g., 'PERSON', 'ORGANIZATION', 'GPE')
        entity_label = subtree.label()
        # Reconstruct the multi-word entity text from the tree leaves
        entity_text = " ".join([token for token, pos in subtree.leaves()])

        print(f"Entity: {entity_text} | Label: {entity_label}")
```
